```python
#data being provided to us is 1 sample/second
#we need to convert it in 1 sample/minute
#1 minute = 60 secs
#then total samples/60= our desired output
#in pandas this can be done using T or Min alias
import pandas as pd
from pandas import read_csv
from datetime import datetime
details = pd.read_csv('details.csv', parse_dates=['Absolute Time'])
detailvol = pd.read_csv('detailVol.csv', parse_dates=['Realtime'])
detailtemp = pd.read_csv('detailTemp.csv',parse_dates=['Realtime'])
```

```python
#creating a functon for dataset that contains Real Time
#function accepts a dataframe as a parameter
def downsampleRT(dataset):
    dataset.drop_duplicates(subset=['Realtime'], inplace=True)
    output = dataset.set_index('Realtime').resample('T').mean()
    return output
```

```python
#creating a functon for dataset that contains Absolute Time
#function accepts a dataframe as a parameter
def downsampleAT(dataset):
    dataset.drop_duplicates(subset=['Absolute Time'], inplace=True)
    output = dataset.set_index('Absolute Time').resample('T').mean()
    return output
```

```python
#calling the function
#output is stored in the variables
detailDownsampled=downsampleRT(detailvol)
detailVolDownsampled=downsampleRT(detailtemp)
detailTempDownsampled=downsampleAT(details)
#output dataframe is exported to .csv files
detailDownsampled.to_csv("detailDownsampled.csv")
detailVolDownsampled.to_csv("detailVolDownsampled.csv")
detailTempDownsampled.to_csv("detailTempDownsampled.csv")
```
